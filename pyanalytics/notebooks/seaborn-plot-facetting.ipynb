```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Load the dataset
df_tips = sns.load_dataset("tips")
```

```python
# Explore the dataset
print(df_tips.info())
```

```python
print(df_tips.head())
```

```python
print(df_tips.describe())
```

```python
# Create a facetgrid
fg = sns.FacetGrid(df_tips, col="day", row="sex")
```

```python
# Plot scatter viz of bill vs tip
fg.map(plt.scatter, "total_bill", "tip", alpha=0.5)
```

```python
# Add titles for each subplot
fg.set_titles("Day: {col_name} | Gender: {row_name}")
fg.set_axis_labels("Total Bill", "Tip")
```

```python
# Adjust layout
plt.tight_layout()
```

```python
# Final plot
plt.show()
```
