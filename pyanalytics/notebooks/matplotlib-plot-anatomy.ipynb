```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter
```

```python
# Basic Plotting

## Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

## Create the plot
figx, axx = plt.subplots()
axx.plot(x, y)

# Remove the top and right spines (borders)
axx.spines['top'].set_visible(False)
axx.spines['right'].set_visible(False)

# Optionally, remove tick marks
#axx.xaxis.set_ticks_position('bottom')
#axx.yaxis.set_ticks_position('left')

# Display the plot
plt.show()
```

```python
# Data
np.random.seed(999)

X = np.linspace(0.5, 3.5, 100)
Y1 = 3 + np.cos(X)
Y2 = 1 + np.cos(1 + X / 0.75) / 2
Y3 = np.random.uniform(Y1, Y2, len(X))
```

```python
print(f"Count: {len(X)}\n\n{X}\n\n{Y1}")
```

```python
# Initiate the base figure
#fig = plt.figure(figsize=(8, 8))
#ax = fig.add_subplot(1, 1, 1, aspect=1)
fig, ax = plt.subplots(figsize=(8, 8))
```

```python
# Remove the top and right spines (borders)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
fig
```

```python
# Util function to define minor ticks
def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    #return "%.2f" % x
    return f"{x:.2f}"
```

```python
# Set x limits
ax.xaxis.set_major_locator(MultipleLocator(1.000))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))

# Set y limits
ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))

ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))
```

```python
fig
```

```python
# Set x and y limits
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)
```

```python
fig
```

```python
# Set thick parameters
ax.tick_params(which='major', width=1.0)
ax.tick_params(which='major', length=10)

ax.tick_params(which='minor', width=.5, labelsize=10)
ax.tick_params(which='minor', length=5, labelsize=10, labelcolor='0.25')
```

```python
fig
```

```python
# Set the grid
ax.grid(which="major",
        axis = "both",
        linestyle="--",
        linewidth=0.3,
        color='.25',
        zorder=-10)
```

```python
fig
```

```python
# Plot the data - lines
ax.plot(X, Y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10)
ax.plot(X, Y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
```

```python
fig
```

```python
# Plot the data -
ax.plot(X, Y3,
        linewidth=0,
        marker='o',
        markerfacecolor='w',
        markeredgecolor='k')
```

```python
fig
```

```python
# Add title and labels
ax.set_title(label = "Anatomy of a figure",
             loc = "left",
             fontsize=20,
             verticalalignment='bottom')

ax.set_xlabel("X axis label")
ax.set_ylabel("Y axis label")
```

```python
fig
```

```python
# Add legend
ax.legend()
```

```python
# Util function for defining circles
def circle(x, y, radius=0.15):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke

    circle = Circle((x, y), radius,
                    clip_on=False, zorder=10, linewidth=1,
                    edgecolor='black', facecolor=(0, 0, 0, .0125),
                    path_effects=[withStroke(linewidth=5, foreground='w')])

    ax.add_artist(circle)

# Util function for defining texts
def text(x, y, text):
    ax.text(x, y, text,
            backgroundcolor="white",
            ha='center', va='top',
            weight='bold', color='blue')
```

```python
# Draw circles and texts

# Minor tick
circle(0.50, -0.10)
text(0.50, -0.32, "Minor tick label")

# Major tick
circle(-0.03, 4.00)
text(0.03, 3.80, "Major tick")

# Minor tick
circle(0.00, 3.50)
text(0.00, 3.30, "Minor tick")

# Major tick label
circle(-0.15, 3.00)
text(-0.15, 2.80, "Major tick label")

# X Label
circle(1.80, -0.27)
text(1.80, -0.45, "X axis label")

# Y Label
circle(-0.27, 1.80)
text(-0.27, 1.6, "Y axis label")

# Title
circle(1.60, 4.13)
text(1.60, 3.93, "Title")

# Blue plot
circle(1.75, 2.80)
text(1.75, 2.60, "Line\n(line plot)")

# Red plot
circle(1.20, 0.60)
text(1.20, 0.40, "Line\n(line plot)")

# Scatter plot
circle(3.20, 1.75)
text(3.20, 1.55, "Markers\n(scatter plot)")

# Grid
circle(3.00, 3.00)
text(3.00, 2.80, "Grid")

# Legend
circle(3.70, 3.80)
text(3.70, 3.60, "Legend")

# Axes
circle(0.5, 0.5)
text(0.5, 0.3, "Axes")

# Figure
circle(-0.3, 0.65)
text(-0.3, 0.45, "Figure")
```
